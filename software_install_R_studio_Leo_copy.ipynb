```python
%%R

# Utility routine for installing packages
install_if_missing <- function(packages) {
    if (length(setdiff(packages, rownames(installed.packages()))) > 0) {
        install.packages(setdiff(packages, rownames(installed.packages())))
    }
}
```

```python
%%R
install_if_missing(c('tidyverse', 'reticulate', 'pivottabler', 'knitr', 'kableExtra', 'IRdisplay', 'Hmisc', 'ICC', 'ggcorrplot', 'parallel', 'doParallel', 'FactoMineR', 'factoextra', 'devtools', 'EnvStats', 'reticulate','xlsx','ggplot2'))
```
